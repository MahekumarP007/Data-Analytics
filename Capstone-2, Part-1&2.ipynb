```python
### Capstone 2 - Part 1: E-Commerce Customer Segmentation


### Introduction

'''
In the realm of online retail, understanding customer behavior is pivotal for enhancing marketing strategies and boosting sales.
This project focuses on customer segmentation using transactional data from a UK-based retailer. 
By transforming transactional data into a customer-centric dataset and applying the K-means clustering algorithm, 
we try to identify distinct customer groups. This segmentation will enable targeted marketing and personalized recommendations,
thereby increasing marketing efficiency and sales.
'''


## Data Retrieval


# We begin by importing necessary libraries and loading the dataset.

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Loading the dataset
df = pd.read_csv('ecommercedata.csv')

# df = pd.read_csv('C:/Users/kumar/Documents/Untitled Folder/ecommerce_data.csv')

# Display the first few rows
df.head()
print("The First few rows and Columns of the Dataset are",  df.head())

# Display dataset information
print("The Dataset total information can be viewed in Brief", df.info())
df.info()


## Data Preprocessing

# Handling Missing Values

# First, we check for missing values in the dataset
df.isnull().sum()
print("The Missing values after checking are", df.isnull().sum())

# Next, we check for duplicate records.
duplicate_count = df.duplicated().sum()
print(f"Number of duplicate records: {duplicate_count}")

# Handling Outliers

'''
Outliers can skew the analysis, especially in clustering. 
We examine the Quantity and UnitPrice columns for potential outliers.
'''

# Boxplot for Quantity

plt.figure(figsize=(10,5))
sns.boxplot(x=df['Quantity'])
plt.title('Boxplot of Quantity')
plt.show()

# Boxplot for UnitPrice

plt.figure(figsize=(10,5))
sns.boxplot(x=df['UnitPrice'])
plt.title('Boxplot of UnitPrice')
plt.show()



##  Encoding Categorical Variables

# Encoding 'Country' using One-Hot Encoding
df_encoded = pd.get_dummies(df, columns=['Country'], drop_first=True)


# Encoding 'StockCode' using Label Encoding ( Generated by ChatGPT) : 
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df_encoded['StockCode'] = le.fit_transform(df_encoded['StockCode'])

# Dropping 'Description' as it's highly cardinal and may not be useful for clustering
df_encoded.drop('Description', axis=1, inplace=True)

# Display the first few rows of the encoded dataframe
print("The First few rows of the encoded DataFrame are", df_encoded.head())
df_encoded.head()


## Transforming Data Types

# Converting 'InvoiceDate' to datetime
df_encoded['InvoiceDate'] = pd.to_datetime(df_encoded['InvoiceDate'], format='%m/%d/%Y %H:%M')

# Displaying the updated dataframe info
print("The Updated DataFrame Info can be viewed as :", df_encoded.info())
df_encoded.info()

# Calculating Total Spend
df_encoded['TotalSpend'] = df_encoded['Quantity'] * df_encoded['UnitPrice']

# Extracting Invoice Month
df_encoded['InvoiceMonth'] = df_encoded['InvoiceDate'].dt.month

# Assuming the latest InvoiceDate in the sample is the reference point
latest_date = df_encoded['InvoiceDate'].max()


# Group by CustomerID to calculate Recency, Frequency, and Monetary ( Generated by ChatGPT) :
rfm = df_encoded.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (latest_date - x.max()).days,
    'InvoiceNo': 'nunique',
    'TotalSpend': 'sum'
}).reset_index()

rfm.columns = ['CustomerID', 'Recency', 'Frequency', 'Monetary']

# Display the RFM table
rfm.head()
print("The RFM Table can be viewed as : ", rfm.head())


'''
 Plotting the distribution of Recency
'''
plt.figure(figsize=(10,5))
sns.histplot(rfm['Recency'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Recency')
plt.xlabel('Recency (Days)')
plt.ylabel('Number of Customers')
plt.show()

'''
 Plotting the distribution of Frequency
'''
plt.figure(figsize=(10,5))
sns.histplot(rfm['Frequency'], bins=20, kde=True, color='salmon')
plt.title('Distribution of Frequency')
plt.xlabel('Frequency (Number of Purchases)')
plt.ylabel('Number of Customers')
plt.show()

'''
 Plotting the distribution of Monetary
'''
plt.figure(figsize=(10,5))
sns.histplot(rfm['Monetary'], bins=20, kde=True, color='lightgreen')
plt.title('Distribution of Monetary')
plt.xlabel('Monetary (Total Spend)')
plt.ylabel('Number of Customers')
plt.show()

'''Observation: The histograms reveal the spread and central tendency of each RFM metric.
   Most customers have low Recency (recent purchases), varying Frequencies, and a range of Monetary values.
'''

'''
Correlation Analysis

 Analyzing correlations between RFM features helps in understanding relationships.

 Correlation matrix
'''

plt.figure(figsize=(8,6))
sns.heatmap(rfm.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of RFM Features')
plt.show()

'''

Observation: 

Recency and Frequency: Negative correlation (more recent customers purchase more frequently).
Recency and Monetary: Negative correlation.
Frequency and Monetary: Positive correlation.
Note: Actual correlations depend on the dataset.

'''


'''
Customer Segments Visualization

 Visualizing clusters helps in interpreting the segments.

 Scatter plot of Frequency vs Monetary
 
'''

plt.figure(figsize=(10,6))
sns.scatterplot(data=rfm, x='Frequency', y='Monetary', hue='Recency', palette='viridis')
plt.title('Frequency vs Monetary colored by Recency')
plt.xlabel('Frequency (Number of Purchases)')
plt.ylabel('Monetary (Total Spend)')
plt.show()

''' 
Observation: Clusters may emerge where high Frequency and Monetary values indicate valuable customers,
whereas low Recency indicates recent activity.
'''

'''
Conclusion: 

In this phase of the project, we successfully retrieved and preprocessed the e-commerce transactional data. 
Through feature engineering, we derived essential RFM metrics that form the foundation for customer segmentation. 
Exploratory Data Analysis provided insights into customer behaviors and the relationships between different metrics. 
The next steps involve applying K-Means clustering to segment customers, analyzing these clusters,
and developing a recommendation system tailored to each segment's preferences.
'''
```

```python
### Capstone 2 - Part 2 : E-Commerce Customer Segmentation 
### Along with Clustering Algorithms: 


### Introduction

'''
In today's competitive market, understanding customer behavior and preferences is crucial for businesses to design effective marketing strategies,
and improve customer satisfaction. This project aims to segment customers based on their purchasing behavior,
and demographic information using K-Means Clustering. Additionally, a recommendation system is implemented,
to suggest products to customers within the same cluster, helping to increase sales and improve marketing effectiveness.
'''

### Data Retrieval

import pandas as pd

# Loading the dataset

data = pd.read_csv('ecommercedata.csv')

# Displaying the first few rows of the dataset
print(data.head())

# Display dataset info
print(data.info())

# Check for missing values
print(data.isnull().sum())

'''
Summary:
- The dataset consists of customer-related features such as purchase history, demographic information, and spending patterns.
- Understanding the context of each feature is crucial for effective clustering.
'''


### Data Preprocessing


# Dropping duplicates
data = data.drop_duplicates()

# Handling missing values
data = data.dropna(subset=['CustomerID'])

# Converting InvoiceDate to datetime
data['InvoiceDate'] = pd.to_datetime(data['InvoiceDate'], format='%m/%d/%Y %H:%M')

# Filtering out records with Quantity <= 0
data = data[data['Quantity'] > 0]

# Reset index
data = data.reset_index(drop=True)

'''
Explanation:
- `drop_duplicates()` removes duplicate rows.
- `dropna(subset=['CustomerID'])` removes rows with missing `CustomerID`, which are crucial for customer-based analysis.
- `pd.to_datetime()` converts `InvoiceDate` to datetime format for easier manipulation.
- `data[data['Quantity'] > 0]` filters out records with non-positive quantities, as they are likely errors or cancellations.
'''


### Feature Engineering & EDA


import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Feature Engineering
data['TotalPrice'] = data['Quantity'] * data['UnitPrice']
data['InvoiceMonth'] = data['InvoiceDate'].dt.to_period('M')

# Aggregate data to get customer-level features
customer_data = data.groupby('CustomerID').agg({
    'TotalPrice': 'sum',
    'InvoiceNo': 'count',
    'InvoiceMonth': 'max'
}).reset_index()
customer_data.columns = ['CustomerID', 'TotalSpend', 'NumPurchases', 'LastPurchaseMonth']

# Exploratory Data Analysis
plt.figure(figsize=(12, 6))
sns.histplot(customer_data['TotalSpend'], bins=50, kde=True)
plt.title('Distribution of Total Spend')
plt.xlabel('Total Spend')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(12, 6))
sns.boxplot(x='NumPurchases', data=customer_data)
plt.title('Number of Purchases per Customer')
plt.xlabel('Number of Purchases')
plt.show()


'''Explanation:
- `data['TotalPrice']` calculates the total price for each transaction.
- `data['InvoiceMonth']` extracts the month of each invoice.
- Aggregating data by `CustomerID` to get total spend, number of purchases, and last purchase month.
- `sns.histplot()` and `sns.boxplot()` are used for visualizing the distribution of total spend and the number of purchases.
'''


### Customer Segmentation using K-Means Clustering


from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Feature scaling
scaler = StandardScaler()
scaled_features = scaler.fit_transform(customer_data[['TotalSpend', 'NumPurchases']])

# K-Means clustering with explicit n_init parameter
kmeans = KMeans(n_clusters=4, n_init=10, random_state=42)
customer_data['Cluster'] = kmeans.fit_predict(scaled_features)

# Cluster analysis
cluster_summary = customer_data.groupby('Cluster').agg({
    'TotalSpend': ['mean', 'std'],
    'NumPurchases': ['mean', 'std']
}).reset_index()
cluster_summary.columns = ['Cluster', 'MeanTotalSpend', 'StdTotalSpend', 'MeanNumPurchases', 'StdNumPurchases']

print(cluster_summary)

# Visualize clusters
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 8))
sns.scatterplot(x='TotalSpend', y='NumPurchases', hue='Cluster', data=customer_data, palette='viridis')
plt.title('Customer Segmentation')
plt.xlabel('Total Spend')
plt.ylabel('Number of Purchases')
plt.show()




'''
Giving Error 

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Feature scaling
scaler = StandardScaler()
scaled_features = scaler.fit_transform(customer_data[['TotalSpend', 'NumPurchases']])



# K-Means clustering
kmeans = KMeans(n_clusters=4, random_state=42)
customer_data['Cluster'] = kmeans.fit_predict(scaled_features)

# Cluster analysis
cluster_summary = customer_data.groupby('Cluster').agg({
    'TotalSpend': ['mean', 'std'],
    'NumPurchases': ['mean', 'std']
}).reset_index()
cluster_summary.columns = ['Cluster', 'MeanTotalSpend', 'StdTotalSpend', 'MeanNumPurchases', 'StdNumPurchases']

print(cluster_summary)

# Visualize clusters
plt.figure(figsize=(10, 8))
sns.scatterplot(x='TotalSpend', y='NumPurchases', hue='Cluster', data=customer_data, palette='viridis')
plt.title('Customer Segmentation')
plt.xlabel('Total Spend')
plt.ylabel('Number of Purchases')
plt.show()
'''


'''Explanation:
- `StandardScaler()` scales features to standardize the data.
- `KMeans(n_clusters=4)` performs K-Means clustering with 4 clusters (you can use methods like the elbow method to determine the optimal number of clusters).
- `cluster_summary` provides an overview of the clusters, including mean and standard deviation of features.
- `sns.scatterplot()` visualizes the clusters.
'''


### Recommendation System


# Identify top-selling products
top_products = data.groupby('Description').agg({'Quantity': 'sum'}).reset_index()
top_products = top_products.sort_values(by='Quantity', ascending=False)

# Example recommendation for a cluster
def recommend_products(cluster_id):
    cluster_customers = customer_data[customer_data['Cluster'] == cluster_id]['CustomerID']
    cluster_transactions = data[data['CustomerID'].isin(cluster_customers)]
    purchased_products = cluster_transactions['Description'].unique()
    
    recommendations = top_products[~top_products['Description'].isin(purchased_products)]
    return recommendations.head(10)

# Recommend products for Cluster 0
print(recommend_products(0))


'''Explanation:
- Aggregates the total quantity of each product.
- Defines a function to recommend products to customers in a specific cluster based on what they haven't purchased yet.
'''


### Conclusion

'''
In conclusion, this project successfully segmented customers into distinct groups using K-Means Clustering, 
allowing for targeted marketing strategies. The recommendation system provides personalized product suggestions, 
further enhancing customer satisfaction and sales potential. Through careful data preprocessing, feature engineering,
and analysis, the project demonstrates the value of data-driven decision-making in a business context.
'''

```
